These are plots of $(r,\theta)=(p,p)$ for all primes $p$ (up to an arbitrary upper bound) in polar coordinates.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import math
```

A rough prime-finding function:

```python
def get_primes(lower, upper):
    """
    Returns np.array of all primes in give range.
    """
    primes = [2]
    for num in range(lower, upper + 1):    
        if num > 1:
            for i in range(2, math.ceil(np.sqrt(num)) + 1):
                if (num % i) == 0:
                    break
            else:
                primes.append(num)
    return np.array(primes)
```

We find all primes less than 100000.

```python
x = get_primes(0,1000000)
x.shape
```

### Plot of the first 5000 primes

```python
df = pd.DataFrame({"r": x[0:5000], "theta": x[0:5000]})
g = sns.FacetGrid(df, subplot_kws=dict(projection='polar'), height=15, despine=False)
g.map(sns.scatterplot, "theta", "r")
```
